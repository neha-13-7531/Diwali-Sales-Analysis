# EDA (Exploratary Data Analysis

```python
import warnings # to ignore the any type of error 
warnings.filterwarnings('ignore')
```

```python
# count for gender
plt.figure(figsize=(8,4))
ax=sns.countplot(x=df['Gender'],palette='coolwarm')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Distribution")
plt.show()   
```

- Female customers are significantly higher than male customers i.e female count are 7832 and male count are 3407.
- This indicates that women are more actively involved in festive shopping during Diwali.

```python
# count for gender
plt.figure(figsize=(8,4))
ax=sns.countplot(x=df['Marital_Status'],palette='coolwarm')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Marital_Status Distribution")
plt.xlabel("Marital_Status")
plt.ylabel("Distribution")
plt.show()   
```

- The Married(6518) customer are  highly purchesing in diwali as comapare to unmarried (4721) customer.

```python
# count plot for gender vs marital status 
plt.figure(figsize=(8,4))
ax=sns.countplot(data=df,x='Gender',hue='Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Gender vs Marital_Status")
plt.xlabel("Gender")
plt.ylabel(",Marital_Status")
plt.show()   
```

- Married females alone contribute the highest share, nearly 4045% of total customers.
- Unmarried males represent the lowest contribution segment.

```python
# count plot for gender vs zone
plt.figure(figsize=(8,4))
ax=sns.countplot(data=df,x='Gender',hue='Zone',palette='coolwarm')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Gender vs Zone")
plt.xlabel("Gender")
plt.ylabel("Zone")
plt.show()   
```

- In every zone, female customer count is consistently higher than males.
- Certain zones show female dominance above 65%.

```python
# count plot for gender vs state
plt.figure(figsize=(24,20))
ax=sns.countplot(data=df,x='Gender',hue='State',palette='coolwarm')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Gender vs State")
plt.xlabel("Gender")
plt.ylabel("State")
plt.show()   
```

- Top 5 states contribute more than half of total sales volume.

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age,palette='coolwarm')
plt.show()
```

- The 2635 age group contributes the highest, approximately 4550% of total orders.
- Followed by 1825 and 3645 age groups.
- Customers above 45 contribute a relatively smaller share, below 15%.

```python
#  univariate for categorical column
def univariate_categorical(categorical_columns, df):
    for column in categorical_columns:
        plt.figure(figsize=(8,6))
        df[column].value_counts().head().plot(kind='bar')
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.show()
categorical_columns = ['Gender', 'Product_Category', 'Occupation', 'Zone']
univariate_categorical(categorical_columns, df)
```

```python
# Bivariate Analysis for categorical column

cat_col=['Gender','Zone','Occupation']
def bivariate_analysis(n):
    plt.figure(figsize=(30,20))
    for i,col in enumerate(n,1):
        plt.subplot(2,2,i)
        plt.title(f"count of {col}")
        sns.countplot(x=df[col],hue='Marital_Status',data=df)
    plt.show()
bivariate_analysis(cat_col)
```

- IT, Healthcare, and Aviation professionals together account for nearly 50% of total customers.
- Students and government employees show lower purchasing participation.

```python
# correlation matrix
correlation=df[['Age','Marital_Status','Orders','Amount']].corr() 
print(correlation)
plt.figure(figsize=(8,4))
sns.heatmap(correlation,annot=True)
plt.title('Heatmap of correlation')
plt.show()
```

- 1 is show the every column is correalated with each other.
- The order and Age this colun is strongly corelatde to each other.

```python
# pie chart of value_counts() of outcome
zone_counts = df['Zone'].value_counts()
custom_colors = ['#008080', '#66ffff', '#00cccc','#009999']
plt.figure(figsize=(4, 4))
plt.pie(zone_counts, labels = zone_counts.index, autopct='%1.1f%%', colors = custom_colors, startangle = 90)
plt.title('Distribution of Zone ')
plt.axis('equal')  
plt.show()
```

- There are Central zone have highest distributuion i.e 38.2%.
- Then Southern have 24.0%.
- The Western ahve 17.4 distributuion .
- The Eastern have lowest distributuion i.e 7.2%

```python
# pie chart of value_counts() of gender
gender_counts = df['Gender'].value_counts()
custom_colors = ['#FFB347', '#77DD77']
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels = gender_counts.index, autopct='%1.1f%%', colors = custom_colors, startangle = 90)
plt.title('Distribution of Gender ')
plt.axis('equal')  
plt.show()
```
